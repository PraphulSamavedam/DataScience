## What is `matplotlib` python library?
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.[<sup>[1]</sup>](#OfficialWebsite)

`import matplotlib.pyplot`

```python
import pandas as pd
# pandas is imported to read , write and data wrangling of wide variety of data.

import numpy as np
# To create dummy data to start with

import matplotlib.pyplot as plt
# %matplotlib inline
# This ensures that the matplotlib library has been imported and all the plots will be plotted when plt.<function()> has been called.
```

### Basic 2D line plot
Let us create a dummy 2D line plot using matplotlib library. Its a single line code, however for that we require some data which we intent to plot a line chart.

```python
X = np.arange(10,300)
# creates an evenly spaced numpy array starting form 10 and ending in 300
Y = np.arange(410,700)
# creates an evenly spaced numpy array starting form 10 and ending in 300
plt.plot(X,Y)
# Creates a line graph of the X and Y variables.
```

### Adding details to the plot
Plot should be self understandable rather looking at the code, we can add those details. 

```python
# Add x-label to the plot
plt.xlabel('X-label')
# Plot the 2D line  of Y vs X
plt.plot(X,Y)
# Add y-label to the plot/chart
plt.ylabel('Y-label')
# Add title for the plot/chart
plt.title('Y vs X line plot')
```

### Adding style to the plot/chart
The third optional parameter can be leveraged to have the styling to the plot.

```python
"""green dashed line plot with value as 'g--' 
    g => green color
    -- => dashed line
    By default the value of style is 'b-'
"""
plt.plot(X,Y,'g--')
plt.show()
```

Color option in the style can have `'r','g','b'` as its values.    
Some of the different markers are `'o','^','-','--','x','*','s','p','d','h' etc.`    
Others you can explore.

By explicitly mentioning the features, we can have more color options and even have better control on the style features like linewidth, markersize, etc.

```python
data = np.linspace(0.01,3.5,num=25)
plt.plot(data, 3**data,marker='*', c='maroon',linestyle='dashed',linewidth='1', markersize='5')
```

### Plot multiple line plot
After the 3 parameters, we can utilized the same plot line code for different plots. Lets plot the functions with the styles mentioned below on a single plot.
* Linear function plot => green line 
* Exponential function plot => red cross (without line)
* Quadratic function plot => blue dashed line

```python
data = np.linspace(0.01,3.5,num=25)
plt.plot(data,data*3,'g-',data,2**data, 'rx',data,data**2,'b--')
plt.xlabel('20 data points')
plt.ylabel('function values')
plt.title('Function values of various data points')
# Add Legend in the same order as that of the plot.
plt.legend(['linear','exponential','quadratic'])
```

### Can we have a signle line chart different inputs?
Yes, in such case.   
What is mandatory relation between X and Y values for the line chart to be plotted

```python
data1=np.arange(start=0.5,stop=5.5,step=0.01)
data2=np.arange(start=0.2,stop=3.5,step=0.01)
data3=np.arange(start=2,stop=3,step=0.1)
plt.plot(data1,data1*2,'r-',data2,data2**2,'--b',data3, data3%10,'og')
plt.xlabel('Different datapoints')
plt.ylabel('Function values')
plt.title('Different inputs can also be plotted')
plt.show()
```

### Can we save the plot generated?
Yes. Using the savefig command we can save the plots/charts generated. 

```python
plt.plot(data, 3**data,c='r',linestyle='dashed',linewidth='2',marker='o', markersize='5')
plt.savefig('generated/images/different input based plot.jpeg')
```

### Multiple plots for data by splitting the area
Using subplots, we can split the data into subparts.

```python
data1=np.arange(start=0.0,stop=5.0,step=0.5)
plt.subplot(1,2,1)
plt.plot(data1, data1*2, 'g-*')
plt.title('Output=2(Input)')
plt.xlabel('Input')
plt.ylabel('Output')

plt.subplot(1,2,2)
plt.plot(data1, data1*3, 'g-*')
plt.title('Output=3(Input)')
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()

plt.subplot(1,1,1)
plt.plot(data1, data1*5, 'g-*', data1, data1*3, 'r--o')
plt.title('Output=n(Input)')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend(['Output=5*input','Output=3*input'])
plt.show()
```

```python
t = np.arange(0,2*np.pi,np.pi/256)
plt.plot(t, np.sin(t),'r',t,np.cos(t),'#008800')
plt.legend(['sin(t)','cos(t)'])
plt.xlabel('Time(t)')
plt.title('Trigonometric functions')
```

## Scatterplot

```python
data= np.arange(0.0,4*np.pi,1/20)
plt.scatter(data, np.sin(data), c='maroon', marker='*', s=1)
plt.scatter(data, np.cos(data), c='black', marker='p',s=1)
plt.title('Scatter plot')
plt.legend(['sin','cos'])
```

## Barplot

```python
import random
data= np.array([random.gauss(5,1) for i in range(0,20)])
plt.bar(data,data*4.2,align='center')
```

```python
data=np.arange(0.1,8,1)
plt.barh(data,data*4,align='center',color=['k','r'])
```

## Histogram plot

```python
data= np.array([np.random.normal(5,1) for i in range(0,1000)])
plt.subplot(2,1,1)
plt.hist(data)
plt.title("Histogram of generated normal random numbers")
plt.subplot(2,1,2)
bin_count=50
plt.hist(data,bins=bin_count)
plt.title("Histogram with bins={}".format(bin_count))
```

```python
# The histogram distribution shall represent the bell curve more clearly in this example.
data= np.array([random.normalvariate(5,1) for i in range(0,100000)])
plt.subplot(2,1,1)
plt.hist(data)
plt.title("Histogram of generated normal random numbers")
plt.subplot(2,1,2)
bin_count=50
plt.hist(data,bins=bin_count, density=True)
plt.title("Histogram with bins={}".format(bin_count))
```

#### Official example given for numpy.random.normal generation [<sup>2</sup>](#NormalDistributionDocumentation)

```python
import matplotlib.pyplot as plt
mu, sigma = 0, 1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)
count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```

### Boxplot

```python
data = [np.random.normal(0,std,100) for std in range(1,4,1)]
plt.boxplot(data,vert=False, patch_artist=True)
plt.show()
```

## Pie Chart

```python
values=[26,35,64,148]
labels=['Directors','Manager','Banker','Clerk']
explodes=[0,0,0,0]
colors=['grey','lightgreen','skyblue','lightblue']
plt.pie(values,explode=explodes,labels=labels,
    colors=colors,
    autopct='%1.1f%%',
    pctdistance=0.6,
    shadow=True,
    labeldistance=1.1,
    startangle=45,
    normalize=True)
```
